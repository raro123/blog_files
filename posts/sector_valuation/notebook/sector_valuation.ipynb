```python
import pandas as pd
```

```python
ratio_df = pd.read_csv('./data/nse_valuation_ratios.csv')
```

```python
ratio_df
```

```python
sector_selection = ['NIFTY AUTO',
'NIFTY CONSUMER DURABLES',
'NIFTY FINANCIAL SERVICES EX-BANK',
'NIFTY FMCG',
'NIFTY HEALTHCARE',
'NIFTY IT',
'NIFTY METAL',
'NIFTY OIL & GAS',
'NIFTY PHARMA',
'NIFTY PRIVATE BANK',
'NIFTY PSU BANK',
'NIFTY REALTY',
]
```

```python
pe_agg_df = (ratio_df
 .query('index_category=="SECTORAL"')
 .query('date>="2005"')
 .sort_values(['index_name','date'])
 .groupby('index_name')
 .agg({'pe':['min','max','mean','median','last',('95_perc', lambda x: x.quantile(0.95))]})
 .droplevel(0, axis=1)
 .reset_index()
 .query('index_name in @sector_selection')
)
```

```python
pe_agg_df
```

```python
pe_agg_df.to_csv('./data/pe_agg.csv', index=False)
```
